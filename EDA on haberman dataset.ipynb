```python
survive=haber.loc[haber['Survivalstat']==1]
death=haber.loc[haber['Survivalstat']==2]
```

```python
counts, binedges = np.histogram(survive['Age'], bins = 20, density = True)
pdf = counts/sum(counts)
print(pdf);
cdf = np.cumsum(pdf)
plt.plot(binedges[1:],pdf,label = 'Spdf');
plt.plot(binedges[1:], cdf,label = 'Scdf');
plt.legend();

counts, binedges = np.histogram(death['Age'], bins = 20, density = True)
pdf = counts/sum(counts)
print(pdf);
cdf = np.cumsum(pdf)
plt.plot(binedges[1:], pdf,label = 'dpdf');
plt.plot(binedges[1:], cdf,label = 'dcdf');
plt.legend();
plt.show();
```

```python
counts, binedges = np.histogram(survive['operationyear'], bins = 20, density = True)
pdf = counts/sum(counts)
print(pdf);
cdf = np.cumsum(pdf)
plt.plot(binedges[1:],pdf,label = 'Spdf');
plt.plot(binedges[1:], cdf,label = 'Scdf');
plt.legend();

counts, binedges = np.histogram(death['operationyear'], bins = 20, density = True)
pdf = counts/sum(counts)
print(pdf);
cdf = np.cumsum(pdf)
plt.plot(binedges[1:], pdf,label = 'dpdf');
plt.plot(binedges[1:], cdf,label = 'dcdf');
plt.legend();
plt.show();
```

```python
counts, binedges = np.histogram(survive['axillarynodes'], bins = 20, density = True)
pdf = counts/sum(counts)
print(pdf);
cdf = np.cumsum(pdf)
plt.plot(binedges[1:],pdf,label = 'Spdf');
plt.plot(binedges[1:], cdf,label = 'Scdf');
plt.legend();

counts, binedges = np.histogram(death['axillarynodes'], bins = 20, density = True)
pdf = counts/sum(counts)
print(pdf);
cdf = np.cumsum(pdf)
plt.plot(binedges[1:], pdf,label = 'dpdf');
plt.plot(binedges[1:], cdf,label = 'dcdf');
plt.legend();
plt.show();
```
